```python
import pymysql.cursors
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
import connect_to_db as cn
```

```python
def draw_boxplot()
```

```python
# pymysql을 이용한 db connection.
connection = cn.make_connection('connection.csv')

cursors = []
sqls = []
results = []
result_list = []
for i in range(6):
    cursors.append(connection.cursor(pymysql.cursors.DictCursor))

sqls.append('select score from comments where is_score = 1;')
sqls.append('select score from comments where is_score = 2;')
sqls.append('select score from comments where is_score = 3;')
sqls.append('select score from comments where es_score = 1;')
sqls.append('select score from comments where es_score = 2;')
sqls.append('select score from comments where es_score = 3;')

for i in range(6):
    cursors[i].execute(sqls[i])
    # result가 dict 형태로 return.
    results.append(cursors[i].fetchall())

connection.close()

for i in range(6):
    results[i] = pd.DataFrame(results[i])
    result_list.append(list(np.array(results[i]['score'].tolist())))
```

```python
plt.rcParams['figure.figsize'] = [20, 15]
xlabel = ['is_1', 'is_2', 'is_3', 'es_1', 'es_2', 'es_3']

plt.boxplot(result_list)
plt.xticks([1, 2, 3, 4, 5, 6], xlabel, fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('score(vote)', fontsize=16)
plt.yscale('log')
plt.show()
```
