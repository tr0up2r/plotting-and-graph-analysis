```python
import igraph
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from igraph import *
```

```python
print(igraph.__version__)
```

```python
data = pd.read_csv('mentor_mentee_relationship_after_preprocessing.csv')
```

```python
data1 = data[['mentor', 'mentee', 'average_is_score']]
data2 = data[['mentor', 'mentee', 'sum_is_score']]
data3 = data[['mentor', 'mentee', 'average_es_score']]
data4 = data[['mentor', 'mentee', 'sum_es_score']]
```

```python
# 그래프 생성
tuples1 = [tuple(x) for x in data1.values]
tuples2 = [tuple(x) for x in data2.values]
tuples3 = [tuple(x) for x in data3.values]
tuples4 = [tuple(x) for x in data4.values]
g1 = igraph.Graph.TupleList(tuples1, directed=True, edge_attrs=['average_is_score'])
g2 = igraph.Graph.TupleList(tuples2, directed=True, edge_attrs=['sum_is_score'])
g3 = igraph.Graph.TupleList(tuples3, directed=True, edge_attrs=['average_es_score'])
g4 = igraph.Graph.TupleList(tuples4, directed=True, edge_attrs=['sum_es_score'])
```

```python
# graph 1
print("Number of vertices in the graph:", g1.vcount())
print("Number of edges in the graph", g1.ecount())
print("Is the graph directed:", g1.is_directed())
print("Maximum degree in the graph:", g1.maxdegree())
print()

# graph 2 (== graph 1)
print("Number of vertices in the graph:", g2.vcount())
print("Number of edges in the graph", g2.ecount())
print("Is the graph directed:", g2.is_directed())
print("Maximum degree in the graph:", g2.maxdegree())
```

```python
components1 = g1.clusters()
print(len(components1))

# same clusters num.
print(f'g1 : {len(g1.clusters())}')
print(f'g2 : {len(g2.clusters())}')
print(f'g3 : {len(g3.clusters())}')
print(f'g4 : {len(g4.clusters())}')
```

```python
g1.vs[0]["name"]
```

```python
# avg degree
print(f'avg degree : {mean(g1.degree())}')
print(f'avg in-degree : {mean(g1.indegree())}')
print(f'avg out-degree : {mean(g1.outdegree())}')
```

```python
# degree distribution
dd = g1.degree_distribution()
print(dd)
```

```python
g_degree = pd.Series(g1.degree())

hist, bins = np.histogram(g_degree, bins=bins)
logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
plt.hist(g_degree, bins=logbins)

plt.xscale('log')
```

```python
print(f"g1 weight(average_is_score) avg : {mean(g1.es()['average_is_score'])}")
print(f"g2 weight(sum_is_score) avg : {mean(g2.es()['sum_is_score'])}")
print(f"g3 weight(average_es_score) avg : {mean(g3.es()['average_es_score'])}")
print(f"g4 weight(sum_es_score) avg : {mean(g4.es()['sum_es_score'])}")
```

```python
g1_weight = np.array(g1.es()['average_is_score'])
n, bins, patches = plt.hist(g1_weight)
plt.show()
```

```python
g3_weight = np.array(g3.es()['average_es_score'])
n, bins, patches = plt.hist(g3_weight)
plt.show()
```

```python
g2_weight = pd.Series(g2.es()['sum_is_score'])

hist, bins = np.histogram(g2_weight, bins=bins)
logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
plt.hist(g2_weight, bins=logbins)

plt.xscale('log')
```

```python
g4_weight = pd.Series(g4.es()['sum_es_score'])

hist, bins = np.histogram(g4_weight, bins=bins)
logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
plt.hist(g4_weight, bins=logbins)

plt.xscale('log')
```
