```python
def make_node_and_degree_csv(relationship_csv, filename):
    data = pd.read_csv(relationship_csv)
    
    data = data[['mentor', 'mentee', 'average_is_score']]
    tuples = [tuple(x) for x in data.values]
    
    # graph 생성.
    g = igraph.Graph.TupleList(tuples, directed=True, edge_attrs=['average_is_score'])
    
    result_for_csv = []

    num_v = g.vcount()
    for i in range(num_v):
        result_for_csv.append([g.vs[i]["name"], g.vs[i].indegree(), g.vs[i].outdegree()])

    fields = ['node_id', 'indegree', 'outdegree']
    cn.write_csv_for_db_update(filename, fields, result_for_csv)
```

```python
make_node_and_degree_csv('../graph/csv/mentor_mentee_relationship_until_november.csv', '../graph/csv/node_and_degree_until_november.csv')
```

```python
data = pd.read_csv('../graph/csv/mentor_mentee_relationship.csv')
```

```python
data1 = data[['mentor', 'mentee', 'average_is_score']]
data2 = data[['mentor', 'mentee', 'sum_is_score']]
data3 = data[['mentor', 'mentee', 'average_es_score']]
data4 = data[['mentor', 'mentee', 'sum_es_score']]
```

```python
# 그래프 생성
tuples1 = [tuple(x) for x in data1.values]
tuples2 = [tuple(x) for x in data2.values]
tuples3 = [tuple(x) for x in data3.values]
tuples4 = [tuple(x) for x in data4.values]
g1 = igraph.Graph.TupleList(tuples1, directed=True, edge_attrs=['average_is_score'])
g2 = igraph.Graph.TupleList(tuples2, directed=True, edge_attrs=['sum_is_score'])
g3 = igraph.Graph.TupleList(tuples3, directed=True, edge_attrs=['average_es_score'])
g4 = igraph.Graph.TupleList(tuples4, directed=True, edge_attrs=['sum_es_score'])
```

```python
btwn = g1.betweenness(weights='average_is_score')
ntile = np.percentile(btwn, 95)
pruned_vs = g1.vs.select([v for v, b in enumerate(btwn) if b >= ntile])
pruned_graph = g1.subgraph(pruned_vs)

igraph.summary(pruned_graph)
```
