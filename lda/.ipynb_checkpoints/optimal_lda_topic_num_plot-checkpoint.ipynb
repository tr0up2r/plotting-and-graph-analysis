```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../lda/csv/coherences_and_perplexities/community_0_perplexity.csv')
```

```python
data = pd.read_csv('../lda/csv/coherences_and_perplexities/community_0_coherences_perplexities_posts.csv')

topic_nums = list(np.array(data['topic_num'].tolist()))
coherences = list(np.array(data['coherence'].tolist()))
perplexities = list(np.array(data['perplexity'].tolist()))
```

```python
plt.plot(topic_nums, coherences)
plt.show()
```

```python
plt.plot(topic_nums, perplexities)
plt.show()
```
